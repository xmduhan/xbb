```python
import pandas as pd
from pandas import DataFrame,Series
```

```python
data = pd.io.excel.read_excel('data/xbb.xls','Sheet1')
data.head()
#data[u'发动机转速-rpm'].head()
```

```python
x = data[u'发动机转速-rpm'].apply(lambda x:round(x/50)*50)
y = data[u'重档挖沙扭矩-N.m'].apply(lambda x:round(x/30)*30)
x = Series(list(x))
y = Series(list(y))
x.min(),x.max(),y.min(),y.max()
```

```python
df = DataFrame({'x':x,'y':y,'c':[1]*len(x)})
df = df.groupby(['x','y'])['c'].sum()
df = df.fillna(0)
df = df.unstack()
df = df.fillna(0)
df
```

```python
def decode(x):
    if x == 0 :
        return 0
    if x > 0 and x < 10:
        return 1
    if x >= 10 and x < 20:
        return 2
    if x >= 20 and x < 30:
        return 3
    if x >= 30 and x < 40 :
        return 4
    if x >= 40 and x < 50 :
        return 5
    if x >= 50 :
        return 6
df = df.applymap(lambda x: 50 if x > 50 else x) 
df
```

```python
plt.close('all')
fig = plt.figure(figsize=(20, 8))
im = plt.imshow(df, interpolation="none",cmap=plt.get_cmap('Reds'))
plt.colorbar(im, use_gridspec=True)
plt.tight_layout()
```
